```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

dataA = np.random.normal(0, 0.8, 1000)
dataB = np.random.normal(-2, 1, 1000)
dataC = np.random.normal(3, 2, 5000)

xmin=-10.0
xmax=10.0
dx=1.0

#   used density=True for histrogram area normalized to 1.
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=np.arange(xmin,xmax+dx,dx))

plt.hist(dataA, label='$\mu$=0.0, $\sigma$=0.8', **kwargs)
plt.hist(dataB, label='$\mu$=-2.0, $\sigma$=1.0', **kwargs)
plt.hist(dataC, label='$\mu$=3.0, $\sigma$=2.0', **kwargs);

plt.xlim(xmin,xmax)
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('Gaussian Distributions')

plt.legend(loc='upper left')
plt.savefig("fig_hist1D_1a")
plt.show()
```
