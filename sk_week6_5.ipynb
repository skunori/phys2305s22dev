```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt 

x = np.random.normal(0,1,100)
y = np.random.normal(1,2,100)

xmin=-5.0; xmax=5.0; dx=0.1
ymin=-5.0; ymax=5.0; dy=0.1
xbins=np.arange(xmin, xmax+dx, dx)   # add dx for the last point.
ybins=np.arange(ymin, ymax+dy, dy) 

fig1=plt.figure(1)
plt.hist2d(x, y, bins=[xbins,ybins], cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')

fig2=plt.figure(2)
plt.subplot(211)
plt.hist(x, label='projection X', bins=xbins, color='g')
plt.legend()
plt.xlim(xmin,xmax)
plt.subplot(212)
plt.hist(y, label='projection Y', bins=ybins, color='r')
plt.xlim(ymin,ymax)    # xlimit for 1D hist.
plt.legend()

fig1.savefig('fig1_hist2D_1')
fig2.savefig('fig2_hist2D_1')
plt.show()
```
