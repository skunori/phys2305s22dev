```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

dataA = np.random.normal(5, 1.0, 10000)
dataB = np.random.uniform(0.00, 10.0, 10000)
dataC = np.random.exponential(1.0,10000)

xmin=-0.0
xmax=10.0
dx=0.1

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=np.arange(xmin,xmax+dx,dx))

plt.hist(dataA, label='gaussian $\mu$=5.0, $\sigma$=1.0', **kwargs)
plt.hist(dataB, label='uniform', **kwargs)
plt.hist(dataC, label='exponet, scale=1.00', **kwargs);

plt.xlim(xmin,xmax)
plt.xlabel('X')
plt.ylabel('Events')
plt.title('10000 events per data sample')

plt.legend(loc='upper right')
plt.savefig('fig_hist1D_2')
plt.show()
```
