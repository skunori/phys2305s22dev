```python
# code originally from Week2, exampe step 5.
import numpy as np
import matplotlib.pyplot as plt

def myfunction(a):
   b=np.sin(a)+a-2.0
   return b

x=np.arange(-10, 10.0, 1.0)
y=myfunction(x)
print(x)
print(y)

xval=np.array([],dtype='float')  # create an empty array.

n=len(x)    # length of x array...
for i in np.arange(0,n-1,1):
   print('i=',i,'  x=',x[i],'  y=',y[i])
   z=y[i]*y[i+1]
   if z <= 0.0:
      print("crossing point  x=",x[i])
      xval=np.append(xval,[x[i]])

yval=myfunction(xval)

plt.plot(x,y,"b")    
xa=[-10.0,10.0]      # two x-values for y=0
ya=[0.0,0.0]
plt.plot(xa,ya,"r")
plt.plot(xval,yval,'bo')
plt.savefig("A3_1_1.png")
plt.show()
```

```python
# code originally from Week2, exampe step 5.
# replace the function form
# then, save xvalue and yvalue
import numpy as np
import matplotlib.pyplot as plt

def myfunction(a):
    # b=np.sin(a)+a-2.0
    b=np.sin(a)/a-0.05
    return b
 
x=np.arange(-10, 10.0, 0.9)
y=myfunction(x)
print(x)
print(y)

xval=np.array([],dtype='float')  # create an empty array.
yval=np.array([],dtype='float')  # create an empty array.

n=len(x)    # length of x array...
for i in np.arange(0,n-1,1):
   print('i=',i,'  x=',x[i],'  y=',y[i])
   dx=x[i+1]-x[i]
   dy=y[i+1]-y[i]
   slope=dy/dx

   xval=np.append(xval,[x[i]])
   yval=np.append(yval,slope)


plt.plot(x,y,"b")    
plt.plot(xval,yval,"r")    
plt.savefig("week3_hw1.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some fake data.
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)

fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('A tale of 2 subplots')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Damped oscillation')

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('Undamped')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def myfunction(a):
    # b=np.sin(a)+a-2.0
    b=np.sin(a)/a-0.05
    return b
 
x=np.arange(-10, 10.0, 0.9)
y=myfunction(x)
print(x)
print(y)

xval=np.array([],dtype='float')  # create an empty array.
yval=np.array([],dtype='float')  # create an empty array.

n=len(x)    # length of x array...
for i in np.arange(0,n-1,1):
   print('i=',i,'  x=',x[i],'  y=',y[i])
   dx=x[i+1]-x[i]
   dy=y[i+1]-y[i]
   slope=dy/dx

   xval=np.append(xval,[x[i]])
   yval=np.append(yval,slope)

fig, axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')

axs[0].plot(x,y,"b")    
axs[1].plot(xval,yval,"r") 
axs[0].set_ylabel("y(x)")
axs[1].set_ylabel("y'(x)")
axs[1].set_xlabel("x")
plt.savefig("week3_hw1a.png")
plt.show()
```
